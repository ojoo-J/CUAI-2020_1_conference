```python
### 행 통째로 가져오기


for i in tqdm(range(pageNo1), desc='pagenum'):
    url = "http://apis.data.go.kr/9710000/BillInfoService2/getBillInfoList?serviceKey=" + key + "&pageNo=" + str(i+1) + "&numOfRows=1000&mem_name_check=&mem_name=&hj_nm=&ord=&start_ord=&end_ord=&process_num=-&start_process_num=-&end_process_num=-&propose_num=-&start_propose_num=-&end_propose_num=-&start_propose_date=2020-05-30&end_propose_date=2020-08-09&start_committee_dt=&end_committee_dt=&bill_kind_cd=&curr_committee=&proposer_kind_cd=&p_proc_result_cd=&b_proc_result_cd=&bill_name=&gbn=&amendmentyn=&budget=&"
    req = requests.get(url)
    html = req.text
    bs = BeautifulSoup(html, 'html.parser')

    item = bs.find_all('item')

    for i in tqdm(item, desc='itemnum'):

        try:
            BillDF.append([i.find('billno').text, i.find('proposedt').text, i.find('billname').text, i.find('summary').text, i.find('passgubn').text])
        except:
            pass
```

```python
## 겹치지 않는 행만 추출

NewDF = ['a']
NewDF[0] = BillDF[0]

for i in tqdm(range(len(BillDF)-1), desc='billdfnum'):
    if BillDF[i+1] != BillDF[i]:
        NewDF.append(BillDF[i+1])
BillDF = NewDF

BillDF
```
